# Scrapping data from a website using BeautifulSoup and Pandas

Este ejercicio lo realicé viendo un video tutorial de el canal de Alex The Analyst en Youtube.
Utilicé un sitio web diferente, además de hacer una limpieza de datos usando Pandas antes de exportar el dataframe resultante.

```python
# Importamos los módulos necesarios para hacer web scrapping
from bs4 import BeautifulSoup
import requests
```

```python
# Asignamos el url del sitio web y obtenemos el texto en html
url = 'https://es.wikipedia.org/wiki/Anexo:Pel%C3%ADculas_con_las_mayores_recaudaciones'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('table') #Buscamos las tablas en el sitio web
```

```python
soup.find('table', class_ = 'wikitable sortable') #Usamos esta función para buscar la tabla directamente
```

```python
table = soup.find('table', class_ = 'wikitable sortable') #Asignamos la variable table para no escribir todo el código
```

```python
print(table)
```

```python
table.find_all('th')[0:9] #Identificamos los títulos de columna que nos interesan
```

```python
column_titles = table.find_all('th')[1:9] # Asignamos una variable a esos títulos
```

```python
column_table_titles = [title.text.strip() for title in column_titles] # Eliminamos los espacios antes y después del texto
print(column_table_titles)
```

```python
# Importamos Pandas para poder crear un dataframe
import pandas as pd
```

```python
df = pd.DataFrame(columns = column_table_titles)
df
```

```python
column_data = table.find_all('tr')[1:] # Buscamos las filas que nos interesan
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
# Comenzaremos la limpieza de este dataset
```

```python
# Vamos a eliminar caracteres indeseados de Recaudación mundial

df["Recaudación mundial"] = df["Recaudación mundial"].str.replace("&", "")
df["Recaudación mundial"] = df["Recaudación mundial"].str.split('.', n=2).str[0]
df["Recaudación mundial"] = df["Recaudación mundial"].str.lstrip("0")
```

```python
df
```

```python
# Vamos a eliminar caracteres indeseados de Presupuesto

df["Presupuesto"] = df["Presupuesto"].str.replace("&", "")
df["Presupuesto"] = df["Presupuesto"].str.split('.', n=2).str[0]
df["Presupuesto"] = df["Presupuesto"].str.lstrip("0")
```

```python
df
```

```python
# Vamos a eliminar caracteres indeseados de Taquilla (EE.UU.)

df["Taquilla (EE.UU.)"] = df["Taquilla (EE.UU.)"].str.replace("&", "")
df["Taquilla (EE.UU.)"] = df["Taquilla (EE.UU.)"].str.split('.', n=2).str[0]
df["Taquilla (EE.UU.)"] = df["Taquilla (EE.UU.)"].str.lstrip("0")
```

```python
df["Taquilla (EE.UU.)"] = df["Taquilla (EE.UU.)"].str.replace("—", "")
```

```python
df
```

```python
# Vamos a eliminar caracteres indeseados de Taquilla (fuera de EE.UU.)

df["Taquilla (fuera de EE.UU.)"] = df["Taquilla (fuera de EE.UU.)"].str.replace("&", "")
df["Taquilla (fuera de EE.UU.)"] = df["Taquilla (fuera de EE.UU.)"].str.split('.', n=2).str[0]
df["Taquilla (fuera de EE.UU.)"] = df["Taquilla (fuera de EE.UU.)"].str.lstrip("0")
```

```python
df
```

```python
# Convertimos las columnas de presupuestos y recaudación a datos numéricos
df['Recaudación mundial'] = pd.to_numeric(df['Recaudación mundial'], errors='coerce')
df['Taquilla (EE.UU.)'] = pd.to_numeric(df['Taquilla (EE.UU.)'], errors='coerce')
df['Taquilla (fuera de EE.UU.)'] = pd.to_numeric(df['Taquilla (fuera de EE.UU.)'], errors='coerce')
df['Presupuesto'] = pd.to_numeric(df['Presupuesto'], errors='coerce')
```
